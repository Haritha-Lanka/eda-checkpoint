## Visualize

Create any visualizations for your EDA here. Make note in the form of code comments of what your thought process is for your visualizations.

```python
#creating a histogram for the column Social_Responsiveness_Scale
# because it has continuous data on a scale to check the distribution of the variable
asdctdf.Social_Responsiveness_Scale.plot(kind ='hist',grid =True)
plt.show()
```

```python
#creating a histogram for the column Qchat_10_Score
# because it has continuous data on a scale to check the distribution of the variable
asdctdf.Qchat_10_Score.plot(kind ='hist',grid =True)
plt.show()
```

```python
#Distribution of sex variable
sns.countplot(x = 'Sex',data = asdctdf)
plt.show()
```

```python
plt.figure(figsize = (8,10))
#pie chart for Ethnicity variable:
ax = asdctdf.Ethnicity.value_counts().plot(kind ='pie')
ax.set_title("Composition of different ethnics ")
```

```python
#Comparison between Sex and Social_Responsiveness_Scale
asdctdf.groupby('Sex').Social_Responsiveness_Scale.sum().plot(kind ='bar')
plt.show()
```

```python
plt.figure(figsize =(9,9))
asdctdf.groupby('Ethnicity').A10_Autism_Spectrum_Quotient.sum().plot(kind ='line')
plt.show()
```
